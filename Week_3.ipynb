```python
# Random student generator   
import random
import csv

ranListOfNames = ["Albert", "Brian", "Carsten", "Danny", "Frederik", "Anna", "Birte", "Carinna", "Dorte", "Frederikke"]

ranListOfImage_urls = ["HappyImage", "SadImage", "PuppyImage", "AnotherImage", "HelloWorldImage"]

ranListOfGrades = ["A+", "A", "A-", "B+", "B", "B-", "C+", "C", "C-"]

ranGender = ["Male", "Female"]

# Student class
class Student():
    
    def __init__(self, name, gender, data_sheet, image_url):
        self.name = name
        self.gender = gender
        self.data_sheet = data_sheet
        self.image_url = image_url
        
    def __repr__(self):
        return 'Student: {} {} {} {}'.format(self.name, self.gender, self.data_sheet, self.image_url)
    
    def get_avg_grade(listOfGrades):
        avg_value = sum(listOfGrades) / len(listOfGrades)
        
# DataSheet class
class DataSheet():    
    
    def __init__(self, courses=[]):
        ranETCS = random.randrange(0, 101)
        ranGrade = str(random.choice(ranListOfGrades))
        self.courses = [Course("Python", "Thomas's Zoom", "Thomas", ranETCS, ranGrade),
                       Course("Security", "Dainels Zoom", "Daniel", ranETCS, ranGrade),
                       Course("Unity", "Jespers Zoom", "Jesper", ranETCS, ranGrade)]
    
    def __repr__(self):
        return 'DataSheet: {}'.format(self.courses)
    
    def get_grades_as_list(self):
        grades = []
        for course in self.courses:
            grades.append(course# mangler .grade - variabel eller lign.)
            return grades

# Course class
class Course():
    
    def __init__(self, name, classroom, teacher, ETCS=None, grade=None):
        self.name = name
        self.classroom = classroom
        self.teacher = teacher
        self.ETCS = ETCS
        self.grade = grade
    
    def __repr__(self):
        return 'Course: {} {} {} {} {}'.format(self.name, self.classroom, self.teacher, self.ETCS, self.grade)

def rngStudent():
    studentAdded = []
    ranNum = random.randrange(1, 25)
    print(ranNum, "Students got added:")
    for n in range(ranNum):
        ranETCS = random.randrange(0, 101)
        ranGrade = str(random.choice(ranListOfGrades))
        studentAdded.append(Student(random.choice(ranListOfNames), 
                                    random.choice(ranGender),
                                    random.choice(DataSheet().courses),
                                    random.choice(ranListOfImage_urls)))
    return studentAdded

def write_list_to_CSV_file(output_file, lst):
    with open(output_file, 'w') as file_object:
        writer = csv.writer(file_object)
        for line in lst:
            writer.writerows(str(lst)) 
            
write_list_to_CSV_file("List_of_students.csv", rngStudent())

print(DataSheet().get_grades_as_list())
rngStudent()
```
